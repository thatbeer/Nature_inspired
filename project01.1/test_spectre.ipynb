```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from run import run_genetic_algorithm
```

### EXP01

```python
burma_conf = './conf/burma01.yaml'
brazil_conf = './conf/brazil01.yaml'

conf = [burma_conf, brazil_conf]

for c in conf:
    run_genetic_algorithm(c)
```

```python
import os

dst_path = os.path.abspath('dst')
burma = os.path.join(dst_path,'burma')
brazil = os.path.join(dst_path,'brazil')
file_name = "test_template.csv"
```

```python
df1 = pd.read_csv(os.path.join(burma,file_name))
df2 = pd.read_csv(os.path.join(brazil,file_name))
df1.columns
```

```python
fig , ax = plt.subplots(2,2,sharex=True,sharey=False)
ax[0][0].plot(df1.trial,df1.best_fitness)
ax[0][1].plot(df1.trial,df1.best_fitness)
ax[1][0].plot(df2.trial,df2.avg_fitness)
ax[1][1].plot(df2.trial,df2.avg_fitness)
```

```python
sns.lineplot(x=df1.trial,y=df1.best_fitness)
sns.lineplot(x=df1.trial,y=df1.avg_fitness)
```

```python
sns.lineplot(x=df2.trial,y=df2.best_fitness)
sns.lineplot(x=df2.trial,y=df2.avg_fitness)
```
