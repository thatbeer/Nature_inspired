test1 : mutate with single swap

```python
functions_data = {
    "crossover_fn": "point_crossover",
    "mutate_fn": "mutate", #mutate(1swap), inversion, scramble, mutli_mutate
    "replace_fn": "replace_weakest",
}

cfg = create_config(
    name="test_mutate01",
    path="./data/burma14.xml", 
    file_name="burma_mutate01",
    num_trial=30, # ignored by Sole_Exp
    parameters=parameters_data, 
    functions=functions_data)

cfg2 = create_config(
    name="test_mutate01",
    path="./data/brazil58.xml", 
    file_name="burma_mutate01",
    num_trial=30, # ignored by Sole_Exp, used by search_trial
    parameters=parameters_data, 
    functions=functions_data)

df01 = search_trail(cfg)
df02 = search_trail(cfg2)
gf01 = search_sole(cfg)
gf02 = search_sole(cfg2)

gens_plot(gf01)
gens_plot(gf02)
trial_plot(df01)
trial_plot(df02)
```

```python
df01.to_csv('./mut_resources/burma_test1.csv')
df02.to_csv('./mut_resources/brazil_test1.csv')
gf01.to_csv('./mut_resources/burma_gent1.csv')
gf02.to_csv('./mut_resources/brazil_gent1.csv')
```
